```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
data = pd.read_excel('inmuebles2.xlsx')
```

```python
#Convertir DataFrame a CSV
data.to_csv("inmuebles2.csv")
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('inmuebles2.csv') 
data
```

```python
#Eliminar primera fila ya que afecta los nombres de las columnas
data1= pd.read_csv('inmuebles2.csv', skiprows=1, low_memory=False)
data1
```

```python
#Se eliminan columnas por ser repetitivas
data2= data1.drop(['0','Referencia'], axis=1)
data2
```

```python
#Identificar valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data3=data2.copy()
```

```python
#Sustituir valores nulos con promedio o media
data3["Superficie"]=data3["Superficie"].fillna(round(data2["Superficie"].mean(),1))
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con mediana en la columna cuantitativa de Precio Venta
data3["Precio Venta"]=data3["Precio Venta"].fillna(round(data2["Precio Venta"].median(),1))
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un string específico
data3["Fecha Alta"]= data3["Fecha Alta"].fillna("2007-04-29 00:00:00")
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
data3["Provincia"]=data3["Provincia"].fillna(method="ffill")
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data3["Tipo"]=data3["Tipo"].fillna(method="bfill")
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por un string específico
data3["Operación"]= data3["Operación"].fillna("Alquiler")
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Se eliminan las columnas con nulos ya que son una gran proporción
data3=data3.dropna(1)
data3
```

```python
#Verificar eliminación de valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Realizamos histograma de variables cuantitativas
plt.hist(x=data3["Tipo"], color="blue", rwidth=0.70)
plt.title("Histograma de Tipo")
plt.xlabel("Tipo")
plt.ylabel("Frecuencia")

plt.show()
```
